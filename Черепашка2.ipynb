# Упражнение 2

```python
import turtle as t
from math import sqrt
t.speed(0)
t.pencolor('blue')
def zero(x,y):
    t.penup()
    t.goto(x,y)
    t.pendown()
    t.forward(50)
    t.right(90)
    t.forward(100)
    t.right(90)
    t.forward(50)
    t.right(90)
    t.forward(100)
    t.up()
    t.home()
    t.down()

def one(x,y):
    t.penup()
    t.goto(x,y-50)
    t.pendown()
    t.left(45)
    t.forward(50*sqrt(2))
    t.right(135)
    t.forward(100)
    t.up()
    t.home()
    t.down()
    
def two(x,y):
    t.penup()
    t.goto(x,y)
    t.pendown()
    t.forward(50)
    t.right(90)
    t.forward(50)
    t.right(45)
    t.forward(50*sqrt(2))
    t.left(135)
    t.forward(50)
    t.up()
    t.home()
    t.down()
    
def three(x,y):
    t.penup()
    t.goto(x,y)
    t.pendown()
    t.forward(50)
    t.right(135)
    t.forward(50*sqrt(2))
    t.left(135)
    t.forward(50)
    t.right(135)
    t.forward(50*sqrt(2))
    t.up()
    t.home()
    t.down()
def four(x,y):
    t.penup()
    t.goto(x,y)
    t.pendown()
    t.right(90)
    t.forward(50)
    t.left(90)
    t.forward(50)
    t.left(90)
    t.forward(50)
    t.right(180)
    t.forward(100)
    t.up()
    t.home()
    t.down()
def five(x,y):
    t.penup()
    t.goto(x+50,y)
    t.pendown()
    t.left(180)
    t.forward(50)
    t.left(90)
    t.forward(50)
    t.left(90)
    t.forward(50)
    t.right(90)
    t.forward(50)
    t.right(90)
    t.forward(50)
    t.up()
    t.home()
    t.down()

def six(x,y):
    t.penup()
    t.goto(x+50,y)
    t.pendown()
    t.right(135)
    t.forward(50*sqrt(2))
    t.left(45)
    t.forward(50)
    t.left(90)
    t.forward(50)
    t.left(90)
    t.forward(50)
    t.left(90)
    t.forward(50)
    t.up()
    t.home()
    t.down()
def seven(x,y):
    t.penup()
    t.goto(x,y)
    t.pendown()
    t.forward(50)
    t.right(135)
    t.forward(50*sqrt(2))
    t.left(45)
    t.forward(50)
    t.up()
    t.home()
    t.down()
def eight(x,y):
    t.penup()
    t.goto(x,y)
    t.pendown()
    for i in range(4):
        t.forward(50)
        t.right(90)
    t.goto(x,y-50)
    for i in range(4):
        t.forward(50)
        t.right(90)
    t.up()
    t.home()
    t.down()
def nine(x,y):
    t.penup()
    t.goto(x,y)
    t.pendown()
    for i in range(4):
        t.forward(50)
        t.right(90)
    t.up()    
    t.goto(x+50,y-50)    
    t.down()
    t.right(135)
    t.forward(50*sqrt(2))
    t.up()
    t.home()
    t.down()
    
one(-150,50)
four(-90,50)
one(-30,50)
seven(30,50)
zero(90,50)
zero(150,50)

t.exitonclick()
```

# Упражнение 4

```python
import turtle
import math
turtle.turtlesize(0.5)

turtle.speed(0)
turtle.shape('circle')
x=-300
y=0
vx=20
vy=60
ay=-10
ry=0.98
rx=0.98
k=0.08
dt=0.05
T=float(input())
turtle.penup()
turtle.goto(-300,0)
turtle.pendown()
turtle.goto(400,0)
turtle.penup()
turtle.goto(-300,0)
turtle.pendown()
for i in range(math.ceil(T/dt)):
    if y<0:
        vy=-ry*vy
        vx=rx*vx
    x+=vx*dt
    y+=vy*dt
    vy+=ay*dt-k*vy*dt
    vx+=-k*vx*dt
    turtle.goto(x,y)
turtle.exitonclick() 

    
 
```

# Упражнение 5

```python
from random import randint
import turtle as t
from math import sqrt

t.speed(0)
t.resizemode("user")
t.shapesize(0.7, 0.7, 1)
t.up()
t.goto(-300,-300)
t.down()
for i in range(4):
    t.forward(600)
    t.left(90)
number_of_turtles = 2
steps_of_time_number = 1000
G=-10000000
g=-1000
X=list()
Y=list()
Vx=list()
Vy=list()
Fx=list()
Fy=list()
pool = [t.Turtle(shape='circle') for i in range(number_of_turtles)]
for unit in pool:
    x=randint(-300,300)
    y=randint(-300,300)
    vx=randint(-200,200)
    vy=randint(-200,200)
    X.append(x)
    Y.append(y)
    Vx.append(vx)
    Vy.append(vy)
    unit.penup()
    unit.speed(0)
    unit.goto(x,y)
    unit.shapesize(0.7, 0.7, 1)

for i in range(steps_of_time_number):
    for j in range(len(pool)):
        ax=0
        ay=0
        for k in range(len(pool)):
            if k==j:
                continue
            r=sqrt((X[j]-X[k])**2+(Y[j]-Y[k])**2) 
            a=G/r**2
            ax+=a*(X[j]-X[k])/r
            ay+=a*(Y[j]-Y[k])/r
        dt=0.03
        if Y[j]>=298 or Y[j]<=-298:
            Vy[j]=-Vy[j]
            Y[j]
        if X[j]>=298 or X[j]<=-298:
            Vx[j]=-Vx[j]
            
        Vx[j]+=ax*dt
        Vy[j]+=ay*dt    
        X[j]+=Vx[j]*dt
        Y[j]+=Vy[j]*dt
        pool[j].goto(X[j],Y[j])
t.exitonclick()        
```
